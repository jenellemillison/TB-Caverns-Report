# Set File Paths

Based on your google drive file path setup, indicate the folder that holds the .nii.gz data files and which file to save the extracted JPGs to.

```python
nii_gz_folder_path = 'drive/MyDrive/Colab Notebooks/Artificial Intelligence/TB_Caverns_Data/' #@param 
save_to_folder_path = "drive/MyDrive/Colab Notebooks/Artificial Intelligence/TB_Caverns_Data/train_jpgs/"  #@param "drive/MyDrive/Colab Notebooks/Artificial Intelligence/TB_Caverns_Data/train_jpgs/"
```

```python
for file in os.listdir(nii_gz_folder_path):
  save_name = ''
  if file.endswith(".nii.gz"):
    file_path = nii_gz_folder_path + file
    save_name = file.split('.')
    save_name[0] += '.jpg'
    save_name = save_name[0]
    print('Saving Image with Name:', save_name, 'Retrieved From:', file_path, 'Saving To:', save_to_folder_path)
    !med2image -i '$file_path' -d '$save_to_folder_path' -o '$save_name' -s -1
```

```python
save_to_nested_folders_path = "drive/MyDrive/Colab Notebooks/Artificial Intelligence/TB_Caverns_Data/train_jpgs_folders/" #@param 
```

Make the Folders for the nested folder data

```python
for file in os.listdir(nii_gz_folder_path):
  if file.endswith(".nii.gz"):
    lung_name = file.split('.')
    print(lung_name)
    new_folder = os.path.join(save_to_nested_folders_path, lung_name[0])
    !mkdir '$new_folder'
```

Take data from saved folder above and sort into folders.

```python
for file in os.listdir(save_to_folder_path):
  lung_name = file.split('-')
  print(lung_name[0])
  source = os.path.join(save_to_folder_path, file)
  dest_folder = os.path.join(save_to_nested_folders_path, lung_name[0])
  !cp '$source' '$dest_folder'
```
